```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 6, 200)
y = (x - 2.5) ** 2 - 1

plt.plot(x, y)
plt.show()
```

```python
def J(theta):
    return (theta - 2.5) ** 2 - 1

def dJ(theta):
    return (theta - 2.5) * 2

theta = 0.0
thetas = [theta]
eplison = 1e-8
alpha = 0.1

while True:
    gredient = dJ(theta)
    last_theta = theta
    theta = theta - alpha * gredient
    thetas.append(theta)
    
    if(abs(J(theta) - J(last_theta)) < eplison):
        break
        
plt.plot(x, J(x))
plt.plot(np.array(thetas), J(np.array(thetas)), color="r", marker='+')
plt.show()
```

### 封装成函数

```python
thetas = []

def gd(initial_theta, alpha, eplison = 1e-8):
    theta = initial_theta
    thetas.append(theta)
    
    while True:
        gredient = dJ(theta)
        last_theta = theta
        theta = theta - alpha * gredient
        thetas.append(theta)

        if(abs(J(theta) - J(last_theta)) < eplison):
            break

def plot_thetas():
    plt.plot(x, J(x))
    plt.plot(np.array(thetas), J(np.array(thetas)), color="r", marker='+')
    plt.show()
```

```python
alpha = 0.01
thetas = []
gd(0, alpha)
plot_thetas()
```

```python
# 换成较大的学习率
alpha = 0.8
thetas = []
gd(0, alpha)
plot_thetas()
```
